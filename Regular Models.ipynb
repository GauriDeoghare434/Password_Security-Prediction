```python
import pandas as pd
import torch
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
from transformers import (
    BertTokenizer, RobertaTokenizer, DebertaTokenizer, AlbertTokenizer, 
    DistilBertTokenizer, ElectraTokenizer, XLNetTokenizer, 
    AutoModelForSequenceClassification, Trainer, TrainingArguments
)

# Load Dataset
df = pd.read_csv("data.csv", on_bad_lines='skip')
print("Dataset Loaded Successfully")

# EDA - Password Length Distribution
df['password'] = df['password'].astype(str).fillna("")
df['length'] = df['password'].apply(len)
sns.histplot(df['length'], bins=30, kde=True)
plt.title("Password Length Distribution")
plt.xlabel("Length of Password")
plt.ylabel("Frequency")
plt.show()

# EDA - Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(df['password']))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of Passwords")
plt.show()

# Remove NaN values in 'strength' column before encoding labels
df = df.dropna(subset=['strength'])

# Encode labels (Assuming 1 = Weak, 2 = Medium, 3 = Strong)
label_mapping = {1: 0, 2: 1, 3: 2}
df['strength'] = df['strength'].map(label_mapping)

# Debugging: Check for NaNs
print("Unique Strength Labels:", df['strength'].unique())
print("Missing Values in Strength:", df['strength'].isna().sum())

# Train-Test Split
train_texts, val_texts, train_labels, val_labels = train_test_split(
    df['password'].astype(str).tolist(), df['strength'].tolist(), test_size=0.2, random_state=42
)

# Handle NaN values before converting to int
train_labels = [0 if pd.isna(x) else int(x) for x in train_labels]
val_labels = [0 if pd.isna(x) else int(x) for x in val_labels]

# Tokenization for multiple models
tokenizers = {
    "bert": BertTokenizer.from_pretrained("bert-base-uncased"),
    "roberta": RobertaTokenizer.from_pretrained("roberta-base"),
    "deberta": DebertaTokenizer.from_pretrained("microsoft/deberta-base"),
    "albert": AlbertTokenizer.from_pretrained("albert-base-v2"),
    "distilbert": DistilBertTokenizer.from_pretrained("distilbert-base-uncased"),
    "electra": ElectraTokenizer.from_pretrained("google/electra-base-discriminator"),
    "xlnet": XLNetTokenizer.from_pretrained("xlnet-base-cased")
}

encodings = {}
for model, tokenizer in tokenizers.items():
    print(f"Tokenizing data with {model} tokenizer...")
    encodings[model] = {
        "train": tokenizer(train_texts, truncation=True, padding=True, max_length=32),
        "val": tokenizer(val_texts, truncation=True, padding=True, max_length=32)
    }

# Convert to PyTorch Dataset
class PasswordDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx], dtype=torch.long)  # ✅ Ensure long type for CE loss
        return item

# Training and Evaluation
models = ["bert", "roberta", "deberta", "albert", "distilbert", "electra", "xlnet"]
for model_name in models:
    print(f"\nTraining {model_name.upper()} model...")

    pretrained_model_name = {
        "bert": "bert-base-uncased",
        "roberta": "roberta-base",
        "deberta": "microsoft/deberta-base",
        "albert": "albert-base-v2",
        "distilbert": "distilbert-base-uncased",
        "electra": "google/electra-base-discriminator",
        "xlnet": "xlnet-base-cased"
    }[model_name]

    model = AutoModelForSequenceClassification.from_pretrained(pretrained_model_name, num_labels=3)

    train_dataset = PasswordDataset(encodings[model_name]['train'], train_labels)
    val_dataset = PasswordDataset(encodings[model_name]['val'], val_labels)

    training_args = TrainingArguments(
        output_dir=f"./results_{model_name}",
        evaluation_strategy="epoch",
        save_strategy="epoch",
        per_device_train_batch_size=16,
        per_device_eval_batch_size=16,
        num_train_epochs=3,
        weight_decay=0.01,
        logging_dir=f"./logs_{model_name}",
        logging_steps=10,
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
    )

    trainer.train()
    print(f"Training {model_name.upper()} completed.")

    # Evaluation
    predictions = trainer.predict(val_dataset)
    preds = torch.argmax(torch.tensor(predictions.predictions), dim=1).numpy()
    print(f"Evaluation Report for {model_name.upper()}:")
    print(classification_report(val_labels, preds))
    
    # Confusion Matrix
    cm = confusion_matrix(val_labels, preds)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Weak', 'Medium', 'Strong'], yticklabels=['Weak', 'Medium', 'Strong'])
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"Confusion Matrix for {model_name.upper()}")
    plt.show()

print("All Models Trained and Evaluated Successfully")
```
